```python
import numpy as np

a = 0
b = np.pi
n = 128

print("*********************************************************************************")
print("""Hasil Program untuk Menyelesaian Integrasi Numerik Int x**2 ∗ sin(2x)dx ;
syarat x(0) = 0.0 , x(1) = pi dengan Metode Trapesium dan Simpson 1/3
Dibuat oleh :

\tNama\t\t : Aprilyanto Setiyawan Siburian
\tNIM\t\t     : 24060121120022
\tProg. Studi\t : Informatika""")
print("\n")
print("--------------------------------------------------------------------------------")
print(" | N     |\t H \t    | Int Trapesium\t|  Err IT % |Int Simpson1/3|  Err IS %   |")
print("--------------------------------------------------------------------------------")

for i in range(1,n+1):
    x = np.linspace(a, b, i+1)
    f = x**2 * np.sin(2*x)
    h = (b - a) / i
    
    # Integrasi Numerik dengan Metode Trapesium
    I_trap = (h/2)*(f[0] + 2 * sum(f[1:i]) + f[i])
    err_trap = abs(np.pi - I_trap)/np.pi*100
    
    # Integrasi Numerik dengan Metode Simpson 1/3
    I_simp = (h/3) * (f[0] + 4*sum(f[1:i:2]) + 2*sum(f[:i:2]) + f[i])
    err_simp = abs(np.pi - I_simp)/np.pi*100
    print(f" | {i}\t | {h:.3f}\t| \t{I_trap:.4f}\t\t| {err_trap:.2f} \t|\t{I_simp:.4f}\t   |\t{err_trap:.2f}\t |")
    
print("--------------------------------------------------------------------------------")
```
