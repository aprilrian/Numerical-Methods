```python
import math
import numpy as np
import warnings
import sympy

def bisection(f,a,b,Tol):
    MAX_STEPS = 1000
    
    # initialize
    delta_x = b - a
    c = a + delta_x / 2.0
    c_array = [ c ]

    f_a = f(a)
    f_b = f(b)
    f_c = f(c)

    # Periksa kurung
    if np.sign(f_a) == np.sign(f_b):
        raise ValueError("tidak mengurung: f(a) dan f(b) harus memiliki tanda yang berbeda")

    # Loop sampai kita mencapai TOLERANSI atau kita mengambil MAX_STEPS
    for step in range(1, MAX_STEPS + 1):
        # Periksa toleransi â€“ bisa juga memeriksa ukuran delta_x
        # Kita memeriksa ini terlebih dahulu karena kita atelah menginisiasi nilai-nilai
        # dalam c dan f_c
        if np.abs(f_c) < tol:
            break
        if np.sign(f_a) != np.sign(f_c):
            b = c
            f_b = f_c
        else:
            a = c
            f_a = f_c
        delta_x = b - a
        c = a + delta_x / 2.0
        f_c = f(c)
        c_array.append(c)
        print (f"{str(step)}, x1 = {str(a)}")
        print (f"{str(step)}, xr = {str(b)}")
        print (f"{str(step)}, f(x1) = {str(f_a)}")
        print (f"{str(step)}, delta = {str(delta_x)}")
        print (f"{str(step)}, xm = {str(c)}")
        print (f"{str(step)}, f(xm) = {str(f_c)}")
        print (f"{str(step)}, f(x1)*f(xm) = {str(f(a)*f_c)}")
        print ("")

    if step == MAX_STEPS:
        warnings.warn('Jumlah maksimum langkah yang terlampaui')

    return c, np.array(c_array)

# mengatur fungsi sebagai fungsi lambda inline
f = lambda x: sympy.exp(x)**3 + x**2 + 4*x - 3
# interval awal
a = 1.0
b = -1.0
tol = 1.e-7
# Mencari akar

x_star, x_array = bisection(f, a, b, tol)
print('akar pada x = {}, f(x*) = {}, {} steps'.format(x_star,f(x_star),len(x_array)))
```
