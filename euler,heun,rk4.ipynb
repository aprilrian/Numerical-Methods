```python
import numpy as np 

# Definisi parameter 
f = lambda t, s: t + s
# ODE
x0 = 0
y0 = 0
xn = 0.1
n = 10

# Metode Euler 
def euler(f,x0,y0,xn,n): 
    h = (xn-x0)/n # ukuran langkah 
    t = np.arange(x0, xn + h, h) # grid Numerik 
    s0 = y0 # kondisi awal
    
# Metode Euler 
    s = np.zeros(len(t)) 
    s[0] = s0
    
    for i in range(0, len(t) - 1): 
        s[i + 1] = s[i] + h*f(t[i], s[i]) 
        t[i + 1] = t[i] + h
    return t, s

t_eu,s_eu = euler(f,x0,y0,xn,n)

# Metode Heun
def heun(f,x0,y0,xn,n):
    h = (xn-x0)/n # ukuran langkah
    t = np.arange(x0,xn+h,h) # grid Numerik
    s0 = y0 # kondisi awal
    
    # Metode heun
    s = np.zeros(len(t))
    s[0] = s0
    
    for i in range(0, len(t) - 1):
        k1 = h*f(t[i], s[i])
        k2 = h*f(t[i]+h, s[i]+k1)
        s[i + 1] = s[i] + 0.5*(k1+k2)
        t[i+1] = t[i]+h
    return t, s

t_heun,s_heun = heun(f,x0,y0,xn,n)

# Metode Runge Kutta orde 4
def rungekutta4(f,x0,y0,xn,n):
    h = (xn-x0)/n # ukuran langkah
    t = np.arange(x0,xn+h,h) # grid Numerik
    s0 = y0 # kondisi awal
    
    # Metode Runge Kutta orde 4
    s = np.zeros(len(t))
    s[0] = s0
    
    for i in range(0, len(t) - 1):
        k1 = h*f(t[i], s[i])
        k2 = h*f(t[i]+ 0.5*h, s[i]+ 0.5*k1)
        k3 = h*f(t[i]+ 0.5*h, s[i]+ 0.5*k2)
        k4 = h*f(t[i]+ h, s[i]+k3)
        s[i + 1] = s[i] + (1/6)*(k1 + 2*k2 + 2*k3+ k4)
        t[i+1] = t[i]+h
    return t, s

t_rk4,s_rk4 = rungekutta4(f,x0,y0,xn,n) 

print("*********************************************************************************************************************************")
print("""Program untuk Menyelesaian Persamaan Differensial
\t\tdy/dx = z+y ; syarat y(0) = 0
dengan Metode Euler, Heun, dan Runge Kutta4
Dibuat oleh :
\tNama\t         : Aprilyanto Setiyawan Siburian
\tNIM\t         : 24060121120022
\tProg. Studi\t : Informatika""")
print("*********************************************************************************************************************************\n")
print("Solusi PDB dy/dx = ùë• + ùë¶, ùë¶(0) = 0")
print("---------------------------------------------------------------------------------------------------------------------------------")
print(" | x\t| h\t| y_Analitik\t|  y_Euler\t| Error %\t| y_Heun\t| Error %\t| y_RungeKutta4\t| Error %\t|")
print("---------------------------------------------------------------------------------------------------------------------------------")

sol_euler = s_eu[-1]
sol_heun = s_heun[-1]
sol_rk4 = s_rk4[-1]
sol_exact = (2.718281828459045**(0.1))-(0.1)-1 
galat_eu = abs((sol_exact-sol_euler)/sol_exact)*100 
galat_heun = abs((sol_exact-sol_heun)/sol_exact)*100 
galat_rk4 = abs((sol_exact-sol_rk4)/sol_exact)*100 

print(f" | 0.1\t| 0.01\t| {sol_exact:.10f}\t| {sol_euler:.10f}\t| {galat_eu:.10f}\t| {sol_heun:.10f}\t| {galat_heun:.10f}\t| {sol_rk4:.10f}\t| {galat_rk4:.10f}\t|")
print("---------------------------------------------------------------------------------------------------------------------------------")
```
