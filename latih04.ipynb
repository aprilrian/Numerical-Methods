```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('seaborn-poster')

%matplotlib inline

def divided_diff(x, y):
    '''
    fungsi untuk menghitung tabel beda hingga
    '''
    n = len(y)
    coef = np.zeros([n, n])
    
    # Kolom pertama adalah y
    coef[:,0] = y
    for j in range(1,n):
        for i in range(n-j):
            coef[i][j] = \
            (coef[i+1][j-1] - coef[i][j-1]) / (x[i+j]-x[i])
    return coef

def newton_poly(coef, x_data, x):
    """
    Mengevaluasi Polinomial newton pada x
    """
    n = len(x_data) - 1
    p = coef[n]
    for k in range(1,n+1):
       p = coef[n-k] + (x -x_data[n-k])*p
    return p

x = np.array([0, 1, 2, 3, 4])
y = np.array([1.0000, 0.5403, -0.4161, -0.9900, -0.6536])

# dapatkan coef beda hingga
a_s = divided_diff(x, y)[0, :]

# evaluasi pada titik-titik data baru
x_new = np.arange(0.0, 4.1, .1)
y_new = newton_poly(a_s, x, x_new)

x_new2 = 2.5
y_new2 = newton_poly(a_s, x, x_new2)

plt.figure(figsize = (12, 8))
plt.plot(x, y, 'bo')
plt.plot(x_new, y_new)
plt.plot(x_new2, y_new2, 'bo')
plt.title('Newton Polynomials')
plt.xlabel('Sumbu x')
plt.ylabel('Sumbu y')
plt.grid()
plt.show()
print("Nilai y pada x = 2,5 adalah {}".format(y_new2))
```
