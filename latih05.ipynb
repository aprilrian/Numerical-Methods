```python
import numpy as np
import numpy.polynomial.polynomial as poly
import matplotlib.pyplot as plt

plt.style.use('seaborn-poster')

x = [321, 322.8, 324.2, 325]
y = [2.50651, 2.50893, 2.51081, 2.51188]
P1_coeff = [1476207.55208333,-13668.6961805555, 42.1874999999999, -0.0434027777777777]
P2_coeff = [-6100679.1125542, 56594.1919191928, -175.000000000003, 0.180375180375183]
P3_coeff = [9396236.0491072, -87291.7968750005, 270.312500000001, -0.279017857142859]
P4_coeff = [-4771763.48863632, 44366.3011363632, -137.499999999999, 0.142045454545453]

# get the polynomial function
P1 = poly.Polynomial(P1_coeff)
P2 = poly.Polynomial(P2_coeff)
P3 = poly.Polynomial(P3_coeff)
P4 = poly.Polynomial(P4_coeff)

x_new = np.arange(320.9, 325.1, 0.1)

fig = plt.figure(figsize = (10,8))
plt.plot(x_new, P1(x_new), 'b', label = 'P1')
plt.plot(x_new, P2(x_new), 'r', label = 'P2')
plt.plot(x_new, P3(x_new), 'g', label = 'P3')
plt.plot(x_new, P4(x_new), 'y', label = 'P4')

plt.plot(x, np.ones(len(x)), 'ko', x, np.zeros(len(x)), 'ko')
plt.title('Lagrange Basis Polynomials')
plt.xlabel('Sumbu x')
plt.ylabel('Sumbu y')
plt.grid()
plt.legend()
plt.show()

L = 2.50651*P1 + 2.50893*P2 + 2.51081*P3 + 2.51188*P4
print('L_koef = ',L)

y_new = np.array(L(323.5))
y_new = L(323.5)
print("Nilai dari x = 323.5 adalah {}".format(y_new))
      
fig = plt.figure(figsize = (10,8))
plt.plot(x_new, L(x_new), 'b', x, y, 'ro')
plt.title('Lagrange Polynomial')
plt.grid()
plt.xlabel('Sumbu x')
plt.ylabel('Sumbu y')
plt.show()
```
